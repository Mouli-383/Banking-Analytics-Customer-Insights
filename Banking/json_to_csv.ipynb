```python
import json
import pandas as pd

# Load JSON
with open("mcc_codes.json", "r") as file:
    data = json.load(file)

# Convert dictionary into dataframe
df = pd.DataFrame(
    list(data.items()),
    columns=["mcc_code", "merchant_category"]
)

# Save CSV
df.to_csv(
    "mcc_codes.csv",
    index=False
)

print("MCC Codes CSV created successfully")
```

```python
import json
import pandas as pd

# Load JSON file
with open("train_fraud_labels.json", "r") as file:
    data = json.load(file)

# Convert dictionary into dataframe
df = pd.DataFrame(
    list(data.items()),
    columns=["transaction_id", "fraud_flag"]
)

# Save as CSV
df.to_csv(
    "train_fraud_labels.csv",
    index=False
)

print("Conversion completed successfully")
```

```python
import pandas as pd

# Read original huge CSV
df = pd.read_csv("transactions_data.csv")

# Take random sample
sample_df = df.sample(
    n=200000,
    random_state=42
)

# Save smaller dataset
sample_df.to_csv(
    "transactions_sample.csv",
    index=False
)

print("Sample dataset created")
```
